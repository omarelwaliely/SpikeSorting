```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import random
```

```python
def splitColumns(datafile):
    df = pd.read_csv(datafile, delimiter='	')
    df.columns = ['first','second']
    return df['first'],df['second']
first, second = splitColumns("data.txt")
```

```python
first
```

```python
def extractFeatures(df,windowSize):
    featureList = []
    threshold = 3.5 * df[:499].std(axis=0)
    arr = df.to_numpy()
    i = 0
    while i < len(arr):
        if arr[i] > threshold:
            j = i
            maximum = arr[i]
            maxInd = i
            while(arr[j] > threshold):
                if arr[j] > maximum:
                    maximum = arr[j]
                    maxInd = j
                j+=1
                
            start = int(max(0, maxInd - windowSize // 2))
            end = int(min(len(arr), maxInd + windowSize // 2))
            
            window = arr[int(start):int(end)]
            std = np.std(window)
            diff = np.max(np.abs(np.diff(window)))
            maxPeak = np.max(window)
            minPeak = np.min(window)
            featureList.append({'std': std, 'diff': diff,'peakIndex' : maxInd, 'maxPeak': maxPeak, 'minPeak':minPeak })
            i = end
        else:
            i += 1
    featuresDF = pd.DataFrame(featureList)
    return featuresDF, threshold
```

```python
def plotFeatures(features):
    X = features[['diff', 'std']]
    kmeans = KMeans(n_clusters=3)
    features['cluster'] = kmeans.fit_predict(X)
    plt.scatter(features['std'], features['diff'], c=features['cluster'], cmap='plasma')
    plt.xlabel('Standard Deviation')
    plt.ylabel('Maximum Difference Between Two Succesive Samples')
    plt.title('K-Means Visualization')
    plt.show()
    return features

```

```python
windowSize = 0.002 * 24414
windowSize = int(windowSize)
first = pd.DataFrame()
second = pd.DataFrame()
first, second = splitColumns("Data.txt")
features1, threshold = extractFeatures(first,windowSize)
features2, threshold2 = extractFeatures(second,windowSize)
features1 = plotFeatures(features1)
features2 = plotFeatures(features2)
features1.to_csv("features1.csv",index = False)
features2.to_csv("features2.csv", index = False)  
```
